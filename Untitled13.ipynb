```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wcat = pd.read_csv('/content/WC_AT.csv')
wcat
```

**EDA** of this datasets

```python
wcat.describe()
```

```python
wcat.isnull().sum()
```

```python
plt.boxplot(wcat.Waist)
plt.show()
```

```python
plt.hist(wcat.Waist)
plt.show()
```

```python
sns.distplot(wcat['AT'])
plt.show()
```

```python
plt.plot(wcat.Waist, wcat.AT, 'rs')
plt.xlabel('waist')
plt.ylabel('AT')
plt.title('Scatterplot')
plt.show()
```

```python
wcat.corr()
```

**Building Model**

```python
import statsmodels.formula.api as smf
model = smf.ols('AT~Waist', data=wcat).fit()
model.params
```

```python
model.summary()
```

**Building Predictions**(Predicted Values of AT)

```python
pred = model.predict(wcat.Waist)
pred
```

**Error(Actual-Predicted)**

```python
model.resid
```

```python
rmse = np.sqrt(np.mean((np.array(wcat['AT'])-np.array(pred))**2))
rmse
```

```python
plt.scatter(x= wcat['Waist'], y=wcat['AT'], color='red')
plt.plot(wcat['Waist'], pred, color='yellow')
plt.xlabel('Waist')
plt.ylabel('AT')
```

**Transformation of model(As R2 values is low as 0.67)**

```python
model2 = smf.ols('AT~np.log(Waist)', data=wcat).fit()
model2.summary()
```

```python
predict2 = model2.predict(pd.DataFrame(wcat['Waist']))
predict2
```

```python
model2.resid
```

```python
rmse = np.sqrt(np.mean((np.array(wcat['AT'])-np.array(predict2))**2))
rmse
```

```python
predict2.corr(wcat.AT)
```

```python
plt.scatter(x= wcat['Waist'], y=wcat['AT'], color='green')
plt.plot(wcat['Waist'], predict2, color='blue')
plt.xlabel('Waist')
plt.ylabel('AT')
```

**Building Model3 (Exponential)**

```python
model3 = smf.ols('np.log(AT)~Waist',data=wcat).fit()
model3.summary()
model3.params
```

```python
model3.summary()
```

```python
pred3 = model3.predict(pd.DataFrame(wcat['Waist']))
pred3
```

```python
predict3 = np.exp(pred3)
predict3
```

```python
rmse = np.sqrt(np.mean((np.array(wcat['AT'])-np.array(predict3))**2))
rmse
```

```python
predict3.corr(wcat.AT)
```

```python
plt.scatter(x= wcat['Waist'], y=wcat['AT'], color='green')
plt.plot(wcat['Waist'], predict3, color='blue')
plt.xlabel('Waist')
plt.ylabel('AT')
```

**Building Model no4**

```python
wcat['waist_sq'] = wcat.Waist*wcat.Waist
wcat
```

```python
model4 = smf.ols('np.log(AT)~Waist+waist_sq',data=wcat).fit()
model4.params
```

```python
model4.summary()
```

```python
pred4 = model4.predict(wcat)
predict4 = np.exp(pred4)
predict4
```

```python
rmse = np.sqrt(np.mean((np.array(wcat['AT'])-np.array(predict4))**2))
rmse
```

```python
plt.scatter(x= wcat['Waist'], y=wcat['AT'], color='green')
plt.plot(wcat['Waist'], predict4, color='blue')
plt.xlabel('Waist')
plt.ylabel('AT')
```
